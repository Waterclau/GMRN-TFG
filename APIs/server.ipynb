```python
import os
import subprocess
import sys
from flask import Flask, jsonify

app = Flask(__name__)

@app.route('/ejecutar_script', methods=['GET'])
def ejecutar_script():
    try:
        command = 'python rnn_gan.py --model "small" --datadir "data" --traindir "results\model4-2" --bidirectional_d --learning_rate 0.3 --feed_previous --pretraining_epochs 10 --feature_matching'

        process = subprocess.Popen(command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.STDOUT)

        while True:
            output = process.stdout.readline().decode()
            if output == '' and process.poll() is not None:
                break
            if output:
                print(output.strip())

                if 'Epoch finalized' in output:
                    archivos = os.listdir('results\model4-2\generated_data\out-results')
                    if len(archivos) > 0:
                        ultimo_archivo = archivos[-1]
                        return ultimo_archivo

        return 'Ejecutándose'

    except Exception as e:
        return f'Error al ejecutar el script: {str(e)}'

if __name__ == '__main__':
    app.run()
```

```python
import os
from flask import Flask, send_file, jsonify

app = Flask(__name__)

@app.route('/obtener_midi', methods=['GET'])
def obtener_midi():
    try:
        midi_file = 'one.mid'  # Ruta del archivo MIDI
        attachment_filename = 'one.mid'  # Nombre del archivo adjunto en la respuesta

        return enviar_archivo(midi_file, attachment_filename)

    except Exception as e:
        return f'Error al obtener el archivo MIDI: {str(e)}'

def enviar_archivo(ruta_archivo, nombre_adjunto):
    try:
        # Leer el contenido del archivo
        with open(ruta_archivo, 'rb') as archivo:
            contenido = archivo.read()

        # Configurar las cabeceras de la respuesta
        cabeceras = {
            'Content-Type': 'audio/midi',
            'Content-Disposition': f'attachment; filename="{nombre_adjunto}"'
        }

        # Devolver el archivo en la respuesta
        return contenido, 200, cabeceras

    except Exception as e:
        return f'Error al enviar el archivo: {str(e)}'

if __name__ == '__main__':
    app.run()
```
